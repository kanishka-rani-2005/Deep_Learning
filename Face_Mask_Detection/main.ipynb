```python
import glob
import cv2

base_dir = r'archive\data'
categories = ['with_mask', 'without_mask']
data = []

for category in categories:
    path = base_dir + '\\' + category  
    label = categories.index(category)

    image_paths = []
    for ext in ['jpg', 'jpeg', 'png', 'bmp']:
        image_paths += glob.glob(path + f'\\*.{ext}')

    print(f"Reading from folder: {path}")
    print(f"Found {len(image_paths)} images.")

    for img_path in image_paths:
        img_arr = cv2.imread(img_path)

        if img_arr is None:
            print(f"Couldn't read image: {img_path}")
            continue

        img_arr = cv2.resize(img_arr, (128,128))
        data.append([img_arr, label])
```

```python
print(len(data))
X=[]
Y=[]
for features,label in data:
  X.append(features)
  Y.append(label)
```

Choose randomly 500 samples 

```python
from sklearn.utils import resample
X, Y = resample(X, Y, n_samples=5000, random_state=42)
```

Convert into numpy array

```python
import numpy as np
X=np.array(X)
Y=np.array(Y)
print(X.shape,Y.shape)
```

```python
X=X/255
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

## using CNN-> Maxpooling->Flatten->Dense

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')  # or softmax if multiclass
])
```

```python

model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['acc'])
```

```python
model.fit(X_train,Y_train,epochs=5,validation_data=(X_test,Y_test))
```

Required for face detection


```python
## for face detection
haar=cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
```

DEFINING FUNCTION NEEDED FOR FACE DETECTION , FOR LABEL DRAWING ,FOR MASK DETECION


```python
def detect_face_mask(img):
  img=img/255
  img = np.expand_dims(img, axis=0)
  pred =model.predict(img.reshape(1,128,128,3))[0][0]
  print(pred)
  return pred
def draw_label(img,label,pos,bg=(255,0,0)):
  text_size=cv2.getTextSize(label,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
  end_x=pos[0]+text_size[0][0]+2
  end_y=pos[1]+text_size[0][1]-2
  cv2.rectangle(img,pos,(end_x,end_y),bg,cv2.FILLED)
  cv2.putText(img,label,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1,cv2.LINE_AA)
## return four coordinates around our face
def detect_face(img):
  coords=haar.detectMultiScale(img)
  return coords
```

```python
import cv2

# Assuming 'cap' is initialized elsewhere, e.g., cap = cv2.VideoCapture(0)

while True:
    # ret returns True if the frame was read successfully
    # frame is the image captured from the webcam
    ret, frame = cap.read()  # captures a frame from the webcam.

    if not ret:
        print("Failed to grab frame")
        break

    # Resize the frame to 128x128 for model compatibility
    img = cv2.resize(frame, (128, 128))
    
    # Detect face using a face detection function (detect_face should be defined elsewhere)
    coords = detect_face(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))  # detect face
    
    # Convert four coordinates to a rectangle and draw on the frame
    for x, y, w, h in coords:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 3)

    # Detect mask or no mask
    pred = detect_face_mask(img)
    print(f"Prediction: {pred}")
    if pred>0.5:
        # Draw label for 'Mask'
        draw_label(frame, 'Mask', (10, 30), (0, 255, 0))
    else:
        # Draw label for 'No Mask'
        draw_label(frame, 'No Mask', (10, 30), (0, 0, 255))
    
    # Show image in a window
    cv2.imshow('window', frame)
    
    # Wait for 1 ms for a key press.
    # If the pressed key is 'x', exit the loop and stop the webcam
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
# Close any OpenCV windows that were opened.
cv2.destroyAllWindows()
```

## PREDICTING ON RAW PICTURE

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np

img = load_img('archive/data/without_mask/without_mask_2.jpg', target_size=(128, 128))
img_array = img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

prediction = model.predict(img_array)[0][0]
print("Prediction:", prediction)

if prediction > 0.5:
    print("Wearing Mask")
else:
    print("No Mask")
```
